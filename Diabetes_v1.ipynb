```python
#Function to view the value counts for the whole data
def v_counts(dataframe):
    for i in dataframe :
        print(dataframe[i].value_counts())
        print(dataframe[i].value_counts(normalize=True))
        print("=======================================================================")
```

**Visualization Of [Yes - NO] Columns and their relation with the target**

```python
cols = ['HighBP', 'HighChol', 'CholCheck','Smoker',
       'Stroke', 'HeartDiseaseorAttack', 'PhysActivity', 'Veggies',
       'HvyAlcoholConsump', 'AnyHealthcare', 'NoDocbcCost', 'DiffWalk']
```

```python
#Create a pivot table for satisfaction versus another rating for easy plotting
    
def create_plot_pivot(data2, x_column):
    _df_plot = data2.groupby([x_column, 'Diabetes_012']).size() \
    .reset_index().pivot(columns='Diabetes_012', index=x_column, values=0)
    return _df_plot
```

```python
fig, ax = plt.subplots(3, 4, figsize=(20,20))
axe = ax.ravel()

for i in range(len(cols)):
    create_plot_pivot(data, cols[i]).plot(kind='bar',stacked=True, ax=axe[i])
    axe[i].set_xlabel(cols[i])
    
fig.show()
```

```python
colors = ['#be4d25','#66ccff']
def analyse_dog(var):
    (data.groupby('Diabetes_012')[var]
     .value_counts(1)
     .unstack()
     .iloc[:,::-1]
     .plot(kind='barh',stacked=True,figsize=(10, 2), color=colors, alpha=1)
     .spines[['top', 'right']].set_visible(False))
    plt.legend(['Yes', "No"],bbox_to_anchor=(1, 1, 0, 0),shadow=False, frameon=False)
    plt.yticks(ticks=[0,1,2], labels=['Non-Diabetic', 'Prediabetic', 'Diabetic'])
    plt.tight_layout()
    plt.title(var, fontsize=18)
    plt.show()
```

```python
for var in cols:
    analyse_dog(var)
```

**The feature "Age" and it's relation with the target**

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.boxplot(x = 'Diabetes_class', y = 'Age', data = data)
plt.title('Age vs Diabetes_class')
plt.subplot(1,2,2)
sns.boxplot(x = 'HighBP', y = 'Age', data = data)
plt.title('Age vs HighBloodPressure')
plt.show()
```

we can see that with the Age grows, the frequency of getting Diabetes or High Blood Pressure becomes higher

```python
pd.crosstab(data.Age,data.Diabetes_class).plot(kind="bar",figsize=(12,5))
plt.title('Diabetes Disease Frequency for Ages')
plt.xlabel('Age')
plt.xticks(rotation=0)
plt.ylabel('Frequency')
plt.show()
```

We know that as the age increases, the chances of diabetes also commonly increases. From above we can say, the median of the age of diabetic people is greater than that of non-diabetic people.
